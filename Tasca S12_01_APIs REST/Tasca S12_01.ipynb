1. Consulta l’API pública JSONPlaceholder utilitzant el mètode GET per obtenir:

* Llista de publicacions (/posts)
* Llista d’usuaris (/users)
* Llista de tasques (/todos)

```python
r_get = requests.get('https://jsonplaceholder.typicode.com/')
#Checking status code of the GET and the content
print(f"Status code: {r_get.status_code}, \n{r_get.text}")
```

```python
# /POSTS request
r_post = requests.get('https://jsonplaceholder.typicode.com/posts/')
status_codes['posts'] = r_post.status_code
r_post_data = r_post.json()
data['posts'] = r_post_data
number_records['posts'] = len(r_post_data)
pprint.pprint(r_post_data)
```

```python
# /USERS request
r_users = requests.get('https://jsonplaceholder.typicode.com/users/')
status_codes['users'] = r_users.status_code
r_users_data = r_users.json()
data['users'] = r_users_data
number_records['users'] = len(r_users_data)
pprint.pprint(r_users_data)
```

```python
# /TODOS request
r_todos = requests.get('https://jsonplaceholder.typicode.com/todos/')
status_codes['todos'] = r_todos.status_code
r_todos_data = r_todos.json()
data['todos'] = r_todos_data
number_records['todos'] = len(r_todos_data)
pprint.pprint(r_todos_data)
```

2. Mostra per pantalla:

La quantitat total de cada recurs.
El codi d'estat de cada petició.

```python
# Collecting status codes and number of records of all tested resources
status_code_df = pd.DataFrame.from_dict(status_codes, orient='index', columns=['Status_code'])
number_records_df = pd.DataFrame.from_dict(number_records, orient='index', columns=['Number_records'])
# creating a dataframe with collected information
pd.concat([status_code_df, number_records_df], axis=1)
```

1. Utilitza l’API d’Open Data BCN per:

Cercar un dataset del teu interès mitjançant package_search i package_show.

```python
# Inputs
keyword = 'sanitat'
url = f"https://opendata-ajuntament.barcelona.cat/data/api/3/action/package_search?q={keyword}"

#API request
kw_search = requests.get(url)

data_search = kw_search.json()
#Checking number of fields
print(len(data_search))
# Checking which keys are
print(data_search.keys())
#Cheking number of fields in 'results' from 'result'
print(len(data_search["result"]["results"]))
#checking type of object it is
print(type(data_search["result"]["results"]))

#Visualising the content
for dtset in data_search["result"]["results"]:
    pprint.pprint(dtset)
```

```python
#Saving data in 'items' object
items = data_search["result"]["results"]

#Visualising the content
pprint.pprint(items[0].keys())
```

```python
#Visualising the fields and records inside
for k in items[0].keys():
    print(f" {k}: {items[0][k]}")
```

```python
#There following fields were selected for further decision-making: 'title', 'format', 'id' and 'package_id' in 'resources'
for j in range(len(items)):
    for i in range(len(items[j]["resources"])):
        print(f"{items[j]["title"]} / {items[j]["resources"][i]["format"]} / {items[j]["resources"][i]["id"]} / {items[j]["resources"][i]["package_id"]}")
```

```python
#These data can be used for further identification of the dataset ID as metadata
available_sets_titles = []
available_sets_formats = []
available_sets_resources = []
available_sets_packages = []
for j in range(len(items)):
    for i in range(len(items[j]["resources"])):
        available_sets_titles.append(items[j]["title"].strip())
        available_sets_formats.append(items[j]["resources"][i]["format"].strip())
        available_sets_resources.append(items[j]["resources"][i]["id"].strip())
        available_sets_packages.append(items[j]["resources"][i]["package_id"].strip())

#Creating dataframe of the interested fields
available_sets = pd.DataFrame({
    "Title": available_sets_titles,
    "Resource": available_sets_resources,
    "Package": available_sets_packages,
    "Format": available_sets_formats
})
display(available_sets)
```

2. Dels resultats obtinguts, selecciona’n un que tingui recursos disponibles en CSV o JSON.

```python
#Search the dataframe for the ID of the resource related to Pharmacies
resource_selected = available_sets[(available_sets["Title"].str.contains('pharmac',case=False,  na=False)) & (available_sets["Format"]=='CSV')]
print(resource_selected)
```

3. Amb el resource_id del recurs seleccionat, realitza una consulta mitjançant datastore_search per:

Recuperar almenys 100 registres.

```python
url = f"https://opendata-ajuntament.barcelona.cat/data/api/action/datastore_search?resource_id={resource_selected['Resource'][0]}&limit=100"
farmaco_data = requests.get(url)
print(farmaco_data.status_code)
farmaco_json = farmaco_data.json()
```

```python
#Checking the fields in this resource
for i in range(len(farmaco_json['result'].keys())):
    for k in farmaco_json['result'].keys():
        pprint.pprint(farmaco_json['result'][k], sort_dicts=False)
```

```python
#Checking the keys to identify fields of interest
print(farmaco_json['result'].keys())
#Visualising the content of the 'records'
print(farmaco_json['result']['records'])
#Checking the number of the elements in 'records'
print(len(farmaco_json['result']['records']))
#Checking which fields are inside 'records'
print(farmaco_json['result']['records'][0].keys())
```

4. Converteix els resultats a un DataFrame.

```python
#Create an empty dataframe with column names from the list of fields in 'records'
datacollect = pd.DataFrame(columns=farmaco_json['result']['records'][0].keys())
#Extracting JSON format of each record and saving it in dataframe by concatenation
for i in range(len(farmaco_json['result']['records'])):
    datacollect = pd.concat(
        [datacollect, pd.DataFrame([farmaco_json['result']['records'][i]])],
        ignore_index=True
    )
print(datacollect.head())
```

5. Desa el DataFrame en un fitxer .csv.

```python
#Saving the dataframe into CSV file
datacollect.to_csv("Pharmacies_BCN_extracted.csv", index=False, encoding='utf-8')
```
