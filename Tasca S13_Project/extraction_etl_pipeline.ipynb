```python
########################################
#########
lista_dc_cnam_0226_cI_part = pd.read_excel("raw_data/3-Lista_DC_CNAM-04.02.2026.xls", sheet_name='DC capit. I parţial comp_', dtype=str)

lista_dc_cnam_0226_cI_part = lista_dc_cnam_0226_cI_part[["Grupa maladiilor pentru compensare ", "Cod DCI", "Denumirea Comuna Internationala (DCI)", "Doza în SI MC", "Cod DC", "Denumirea comercială (DC)", "Doza", "Forma farmaceutică", "Divizarea", "Ţara", "Firma producătoare", "Număr de înregistrare", "Data înregistrării", "Cod ATC", "Cod medicament (Catalogul național de prețuri)"]]
lista_dc_cnam_0226_cI_part.columns = lista_dc_cnam_0226_cI_part.columns.str.strip()
lista_dc_cnam_0226_cI_part["is_child_rate"] = False
lista_dc_cnam_0226_cI_part["coverage_type"] = "Partial"

#########
lista_dc_cnam_0226_cI_compl = pd.read_excel("raw_data/3-Lista_DC_CNAM-04.02.2026.xls", sheet_name='DC capit. 1(100%)', dtype=str)

lista_dc_cnam_0226_cI_compl = lista_dc_cnam_0226_cI_compl[["Grupa maladiilor pentru compensare ", "Cod DCI", "Denumirea Comuna Internationala (DCI)", "Cod DC",	"Doza în SI MC", "Denumirea comercială (DC)", "Doza", "Forma farmaceutică", "Divizarea", "Ţara", "Firma producătoare", "Număr de înregistrare",	"Data înregistrării", "Cod ATC", "Cod medicament (Catalogul național de prețuri)"]]
lista_dc_cnam_0226_cI_compl.columns = lista_dc_cnam_0226_cI_compl.columns.str.strip()
lista_dc_cnam_0226_cI_compl["is_child_rate"] = False
lista_dc_cnam_0226_cI_compl["coverage_type"] = "Full"

#########
lista_dc_cnam_0226_analogInsulina_compl = pd.read_excel("raw_data/3-Lista_DC_CNAM-04.02.2026.xls", sheet_name='DC (100%)_Analogi de insulină', dtype=str)

lista_dc_cnam_0226_analogInsulina_compl = lista_dc_cnam_0226_analogInsulina_compl[["Grupa maladiilor pentru compensare ", "Cod DCI",	"Denumirea Comuna Internationala (DCI)", "Doza în SI MC", "Cod DC", "Denumirea comercială (DC)", "Doza", "Forma farmaceutică", "Divizarea", "Ţara", "Firma producătoare", "Număr de înregistrare", "Data înregistrării", "Cod ATC", "Cod medicament (Catalogul național de prețuri)"]]
lista_dc_cnam_0226_analogInsulina_compl.columns = lista_dc_cnam_0226_analogInsulina_compl.columns.str.strip()
lista_dc_cnam_0226_analogInsulina_compl["is_child_rate"] = False
lista_dc_cnam_0226_analogInsulina_compl["coverage_type"] = "Full"

#########
lista_dc_cnam_0226_cII = pd.read_excel("raw_data/3-Lista_DC_CNAM-04.02.2026.xls", sheet_name='DC_capit.II', dtype=str)

lista_dc_cnam_0226_cII = lista_dc_cnam_0226_cII[["Grupa maladiilor pentru compensare ", "Cod DCI", "Denumirea Comuna Internationala (DCI)", "Doza în SI MC", "Cod DC", "Cod DCI_sec", "Cod DC_sec", "Denumirea comercială (DC)", "Doza", "Forma farmaceutică", "Divizarea", "Ţara", "Firma producătoare", "Număr de înregistrare", "Data înregistrării", "Cod ATC", "Cod medicament (Catalogul național de prețuri)"]]
lista_dc_cnam_0226_cII.columns = lista_dc_cnam_0226_cII.columns.str.strip()
lista_dc_cnam_0226_cII["is_child_rate"] = True
lista_dc_cnam_0226_cII["coverage_type"] = "Full"

#########
lista_dc_cnam_0226_paliat = pd.read_excel("raw_data/3-Lista_DC_CNAM-04.02.2026.xls", sheet_name='Îngrijiri Paliative(100%)', dtype=str)

lista_dc_cnam_0226_paliat = lista_dc_cnam_0226_paliat[["Grupa maladiilor pentru compensare ", "Cod DCI", "Denumirea Comuna Internationala (DCI)", "Doza în SI MC", "Cod DC", "Denumirea comercială (DC)", "Doza", "Forma farmaceutică", "Divizarea", "Ţara", "Firma producătoare", "Număr de înregistrare", "Data înregistrării", "Cod ATC", "Cod medicament (Catalogul național de prețuri)"]]
lista_dc_cnam_0226_paliat.columns = lista_dc_cnam_0226_paliat.columns.str.strip()
lista_dc_cnam_0226_paliat["is_child_rate"] = False
lista_dc_cnam_0226_paliat["coverage_type"] = "Full"

#########
lista_dc_cnam_0226 = pd.concat([lista_dc_cnam_0226_cI_part, lista_dc_cnam_0226_cI_compl, lista_dc_cnam_0226_analogInsulina_compl, lista_dc_cnam_0226_cII, lista_dc_cnam_0226_paliat], axis=0, ignore_index=True)

########################################
DCI_1225 = pd.read_excel("raw_data/Anexa-1-la-ordin-ro_decembrie-23.12.25.xlsx", skiprows=3, skipfooter=586,sheet_name='DCI')
DCI_1225.columns = DCI_1225.columns.str.strip()

########################################
# represents structured table with hyerarchical composition --> needs normalisation as child-parent scheme
CIM_1225 = pd.read_excel("raw_data/Anexa-1-la-ordin-ro_decembrie-23.12.25.xlsx", sheet_name='CIM')
CIM_1225.columns = CIM_1225.columns.str.strip()
```

```python
# Saving locally for TRANSFORMATION step
lista_dc_cnam_0226.to_csv("transforming/lista_dc_cnam.csv", header = True, index = False)
DCI_1225.to_csv("transforming/dci.csv", header = True, index = False)
CIM_1225.to_csv("transforming/cim.csv", header = True, index = False)
```
