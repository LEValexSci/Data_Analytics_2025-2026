Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són

- Kelvin (K). 

També existeixen altres unitats com

- Réaumur (°Re).

Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions, de manera que es puguin guardar (recorda que un print() no es pot guardar mai).
Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduides per l’usuari/a no estan en el format adequat.
(EXTRA): Pensa una manera d’emmagatzemar totes les possibles conversions en un sol objecte (Llista? Diccionari? DataFrame?) en comptes d’escriure molts if else en funció de la temperatura d’origen i la temperatura de destí.

```python
#########################################################
# definición de funciones de cálculo de temperatura basadas en el valor Celsius
def converter(C):
    return {
        "K": round(C + 273.15,2),
        "F": round(C * 9/5 + 32, 2),
        "Ra": round(C * 9/5 + 491.67,2),
        "Re": round(C * 4/5, 2),
        "C": C
    }

#########################################################
# definición del diccionario de funciones de conversión de cualquier otra escala a la escala Celsius
XtoC = {
    "K": lambda temp: round(temp - 273.15, 2),
    "F": lambda temp: round((temp - 32) * 5/9, 2),
    "RA": lambda temp: round((temp - 491.67) *5/9, 2),
    "RE": lambda temp: round(temp * 5/4, 2)
}

##################################################
# declaración de la función de conversión de temperatura
def T_convertor():
    while True:
        try:
            temp = float(input("Introduzca el valor numérico de la temperatura."))
            break
        except ValueError:
            print(f"Debe introducir un valor numérico para la temperatura. Por favor, inténtelo de nuevo.")

    while True:
        try:
            scale = str(input("Introducir la escala de temperatura, por ejemplo, C para Celsius, F para Fahrenheit, K para Kelvin, Ra para Rankine, Re para Réaumur.")).upper()
            if scale in {'C', 'K', 'F', 'RA', 'RE'}:
                break
        except ValueError:
            print(f"Se ha seleccionado una escala de temperatura incorrecta. Elija una de la lista: 'C', 'K', 'F', 'Ra', 'Re'")

    if float(temp):
            if scale == "C":
                r = converter(temp)
            else:
                r = converter(XtoC[scale](temp))
                
    return r
```

*3. Comptador de paraules d'un text.*

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casuístiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina és la llargada mitjana de les paraules del text que has escrit? “Hola com va?” hauria de retornar (4+3+2) / 3 = 3



```python
def comptador_paraulas():
    frase = str(input("Por favor, escriba una frase: "))
        
    try:
        if frase and not frase.isspace():

            translator = str.maketrans(string.punctuation, ' ' * len(string.punctuation))
            frase_limpia = frase.translate(translator).split()
            if (len(frase_limpia) != 0):
                w_count = {}
                w_len = {}
                for c in frase_limpia:
                    c = c.lower()
                    if c in w_count:
                        w_count[c] += 1
                    else:
                        w_count[c] = 1
                        w_len[c] = len(c)
        else:
            w_count = "Vacío"
    except:
        print("Inténtalo de nuevo")
    return (w_count)
```

```python
paraulas_calculadas = comptador_paraulas()

# visualisation of words and their counting values
#print(paraulas_calculadas)
pprint.pprint(paraulas_calculadas)
```

```python
#defining the function calculating the average of all words in the frase
def mitjana_paraules(paraulas_calculadas): 
    if paraulas_calculadas != 'Vacío':
        smP = sum(len(k) * v for k, v in paraulas_calculadas.items())
        mediaLong = round(smP/sum(paraulas_calculadas.values()),2)
    else:
        mediaLong = "Vacío"
    return mediaLong
```

```python
#returning average value
print(mitjana_paraules(paraulas_calculadas))
```

*4. Diccionari invers (amb possibilitat de duplicats)*

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviant les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment, juntament amb una llista amb els valors associats a la clau repetida.

Fixa’t en el següents exemples per veure l’output esperat.

```python
def rev_dict(dict):
    rev_keys = [v for v in dict.values()]
    rev_keys_uniq = list(set(rev_keys))
    rev_values = [k for k in dict.keys()]
    rev_dict = dict.fromkeys(rev_keys_uniq)

    for ke in rev_keys_uniq:
        pos = [i for i, x in enumerate(rev_keys) if x == ke]
        val_fin = []
        for p in pos:
            val_fin.append(rev_values[p])

        rev_dict[ke] = val_fin
    for k,v in rev_dict.items():
        if len(v)>1: 
            print(f"Hay diversos valores para {k}: {v}")

    return rev_dict
```

*1. Comptador i endreçador de paraules d'un text.*

El client va quedar partits amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:

```python
def contador_content(file):
    with open(file, 'r', encoding='utf-8') as f:
        content = f.read()
    
    if content and not content.isspace():
        punctuation_symbols = string.punctuation + '¡¿'
        translator = str.maketrans(punctuation_symbols, ' ' * len(punctuation_symbols))
        frase_limpia = content.translate(translator).split()
        if (len(frase_limpia) != 0):
            w_count = {}
            for c in frase_limpia:
                c = c.lower()
                if c in w_count:
                    w_count[c] += 1
                else:
                    w_count[c] = 1
        else:
            w_count = "Empty"
    
    #group by first letter into a dictionary
    fin_libr = {}
    keys_selector = sorted(list(w_count.keys()))
    for k in keys_selector:
        obj = {k: w_count[k]}
        if k[0] in fin_libr:
            fin_libr[k[0]].update(obj)
        else:
            fin_libr[k[0]] = obj

    return fin_libr
```

*2. Conversió de tipus de dades.*

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client:

[ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

imatge 4

```python
def typeCheck(dataList):
    element = []
    for el in dataList:
        if isinstance(el, (list, tuple, set)):
            element.extend(el)
        else:
            element.append(el)

    float_list = []
    nonfloat_list = []
    for el in element:
        try:
            float(el)
            float_list.append(el) # will collect the elements of the list convertible into float, not the result of conversion
            #float_list.append(float(el)) # if the results of conversion are needed
        except:
            nonfloat_list.append(el)
    
    return ({
        'float_convertible': float_list, 
        'float_nonconvertible': nonfloat_list})
```

*1. Generador de Contrasenyes*

Explora el funcionament del mòdul random de la llibreria numpy. 🔗 Random module in NumPy

En aquest punt, el client ha detectat un problema amb les contrasenyes que utilitzen els seus treballadors. Asdf1234, dates d’aniversari o similars. Per a solucionar-ho ens ha encarregat una funció de Python que generi contrasenyes més segures. La funció ha de dependre dels següents paràmetres:

longitud (int): Longitud de la contrasenya
majuscules (bool = True): Si hi ha d’aparèixer majúscules
minuscules (bool = True): Si hi ha d’aparèixer minúscules
numeros (bool = True): Si hi ha d’aparèixer números
signes (bool = False): Si hi ha d’aparèixer caràcters especials (,-$? o similars)
Així doncs, si executem la funció de la següent manera:

Hauríem d’obtenir un output (que hem de poder guardar) de l’estil:

9Er,5Vn8P$

Assegura’t que es compleixin tots els criteris, i que aquestes contrasenyes siguin realment aleatòries.

(EXTRA) Explora com podríem fer que la funció copiés la contrasenya automàticament al porta papers de l’ordinador (com si l’haguéssim seleccionat i fet ctrl+copy).

```python
import subprocess
```

```python
def crear_contrasenya(longitud, majuscules = True, minuscules = True, numeros = True, signes = True, copy = True):
    # determining how many types of symbols required
    nr_req_symb_types = majuscules + minuscules + numeros + signes
    # symbol sets
    capitalSymb = list(string.ascii_uppercase)
    minorSymb = list(string.ascii_lowercase)
    numberSymb = list(string.digits)
    specSymb = list(string.punctuation)

    #define 'password' object to collect all symbols
        # To ensure that password includes at least one element pf each type we randomly select a symbol of each required type
    password = []
    required_symbols = []
    if majuscules:
        password.extend(np.random.choice(capitalSymb))
        required_symbols.extend(capitalSymb)
    if minuscules:
        password.extend(np.random.choice(minorSymb))
        required_symbols.extend(minorSymb)
    if numeros:
        password.extend(np.random.choice(numberSymb))
        required_symbols.extend(numberSymb)
    if signes:
        required_symbols.extend(specSymb)
        password.extend(np.random.choice(specSymb)) 

    # combine first several elements with the generated rest
    password.extend(np.random.choice(required_symbols, longitud - nr_req_symb_types))
    # reordering elements position
    np.random.shuffle(password)
    password =''.join(password)
    
    #copy generated password to memory for pasting
    if copy: subprocess.run(['clip'], input=password.strip().encode('utf-8'), check=True)

    return password
```

```python
def match_statistics(file, orderByColumn = "Points"):
    # loading the file with match results
    partits = pd.read_csv(file, delimiter='\t', header = None, names = ["Team_1", "Score", "Team_2"])
    scores = partits["Score"].str.split("-", expand=True).astype(int)

    # Goals scored
    partits["GoalsScored_T1"] = scores[0]
    partits["GoalsScored_T2"] = scores[1]
    # Goals missed
    partits["GoalsMissed_T1"] = scores[1]
    partits["GoalsMissed_T2"] = scores[0]

    # Evaluating the resuls of the matches 
    Result_T1 = []
    Result_T2 = []
    for r in range(0,len(partits)):
        Result_T1.append("Win") if partits["GoalsScored_T1"][r] > partits["GoalsScored_T2"][r] else Result_T1.append("Lost") if partits["GoalsScored_T1"][r] < partits["GoalsScored_T2"][r] else Result_T1.append("Pair")
        Result_T2.append("Lost") if partits["GoalsScored_T1"][r] > partits["GoalsScored_T2"][r] else Result_T2.append("Win") if partits["GoalsScored_T1"][r] < partits["GoalsScored_T2"][r] else Result_T2.append("Pair")
    partits["Result_T1"] = Result_T1
    partits["Result_T2"] = Result_T2
    
    # Calculating the points received
    Points_T1 = []
    Points_T2 = []
    for r in range(0,len(partits)):
        Points_T1.append(3) if partits["Result_T1"][r] == "Win" else Points_T1.append(0) if partits["Result_T1"][r] == "Lost" else Points_T1.append(1)
        Points_T2.append(3) if partits["Result_T2"][r] == "Win" else Points_T2.append(0) if partits["Result_T2"][r] == "Lost" else Points_T2.append(1)
    partits["Points_T1"] = Points_T1
    partits["Points_T2"] = Points_T2

    resultsSeries_T1 = partits.groupby("Team_1")["Result_T1"].value_counts().reset_index().pivot(index='Team_1', columns='Result_T1', values='count').reset_index()
    resultsSeries_T2 = partits.groupby("Team_2")["Result_T2"].value_counts().reset_index().pivot(index='Team_2', columns='Result_T2', values='count').reset_index()
    resultsSeries = resultsSeries_T1.merge(resultsSeries_T2, left_on="Team_1", right_on="Team_2")
    resultsSeries["Win"] = resultsSeries["Win_x"] + resultsSeries["Win_y"]
    resultsSeries["Pair"] = resultsSeries["Pair_x"] + resultsSeries["Pair_y"]
    resultsSeries["Lost"] = resultsSeries["Lost_x"] + resultsSeries["Lost_y"]

    # Creating final dataframe with all required parameters
    equips_goalsResults = pd.DataFrame({
        "GoalsScored": pd.DataFrame(partits.groupby("Team_1")["GoalsScored_T1"].sum())["GoalsScored_T1"] + pd.DataFrame(partits.groupby("Team_2")["GoalsScored_T2"].sum())["GoalsScored_T2"],
        "GoalsMissed": pd.DataFrame(partits.groupby("Team_1")["GoalsMissed_T1"].sum())["GoalsMissed_T1"] + pd.DataFrame(partits.groupby("Team_2")["GoalsMissed_T2"].sum())["GoalsMissed_T2"],
        "Points": pd.DataFrame(partits.groupby("Team_1")["Points_T1"].sum())["Points_T1"] + pd.DataFrame(partits.groupby("Team_2")["Points_T2"].sum())["Points_T2"]
    })

    equips_stats = resultsSeries.set_index("Team_1")[["Win", "Pair", "Lost"]].merge(equips_goalsResults, left_index=True, right_index=True).sort_values(orderByColumn, ascending=False)
    reportStats = {
        "Goals Scored" : equips_stats["GoalsScored"],
        "Maximal Goals Scored": equips_stats["GoalsScored"].idxmax(),
        "Maximal Goals Missed": equips_stats["GoalsMissed"].idxmax(),
        "General Statistics" : equips_stats
    }
    return reportStats
```
